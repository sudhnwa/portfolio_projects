- <h4>What is overall sales trend ?</h4>

```python
px.line(data_frame=sales_df, x=sales_df['order_date'], y=sales_df['sales'])
```

plotting for sales each day creates a line plot hard to comprehend and analyse, grouping on the month and year and and plotting the sum of of the sales that month might solve the issue.  

```python
sales_df['month_year'] = sales_df['order_date'].apply(lambda x : x.strftime('%Y-%m'))
```

```python
sales_df['month_year']
```

```python
monthly_sales_df = sales_df.groupby('month_year')['sales'].sum()
monthly_sales_df
```

```python
px.line(data_frame=monthly_sales_df, markers=True, labels = {'value' : 'sales', 'month_year' : 'year'})
```

Insights from the visual

From the lineplot we can see that, for each year the sales of first four to five months are comparatively low than the sales in the last four to five months of the year, this indicates a seasonal cycle in the sales.

__

- <h4>Which are the top 10 products by the sales ? </h4>

```python
prod_total_sales_df = pd.DataFrame(sales_df.groupby('product_name')['sales'].sum())
```

```python
prod_total_sales_df.sort_values(by = 'sales', inplace=True, ascending= False)

# # hence top 10 products by sales are 
top_ten_df = prod_total_sales_df[: 10]
# indexing and slicing is used to get the first 10 records in the series

top_ten_df
```

```python
px.bar(data_frame=top_ten_df, x=top_ten_df.index, y=top_ten_df['sales'], color=top_ten_df.index)
```

__

- <h4>Which are the most selling products ? </h4>

```python
prod_total_quant_df = pd.DataFrame(sales_df.groupby('product_name')['quantity'].sum())
prod_total_quant_df
```

```python
prod_total_quant_df.sort_values(by = 'quantity', ascending = False, inplace = True)
top_ten_quant_df = prod_total_quant_df[: 10]
```

```python
top_ten_quant_df
```

```python
px.bar(data_frame=top_ten_quant_df, x=top_ten_quant_df.index, y=top_ten_quant_df['quantity'], color=top_ten_quant_df.index)
```

__

- <h4>What is the most preferred ship mode ? </h4>

```python
ship_mode_count_df = pd.DataFrame(sales_df['ship_mode'].value_counts())
ship_mode_count_df
```

```python
px.bar(ship_mode_count_df, labels = {'index' : 'mode', 'value' : 'count'}, color=ship_mode_count_df.index)
```

- <h4>Which are the most profitable category and sub-category ?</h4>

```python
cat_subcat_profit = pd.DataFrame(sales_df.groupby(['category', 'sub_category'])['profit'].sum())
cat_subcat_profit.sort_values(by = ['category', 'profit'], ascending= False, inplace = True)
cat_subcat_profit
```
